```python
#import sys
#!{sys.executable} -m pip install -r requirements.txt
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import norm
```

```python
## Part A: Simulate and plot Gaussian distribution

# Set up figure and axes for plotting
fig, axs = plt.subplots(3, 3, figsize = (30,30))
# Create counter to track number of iterations and calculate row and col values
k = 0
for i in range(10, 19):
    # Calculate row number
    row = k // 3
    # Calculate column number
    col = k % 3
    
    # Define parameters
    mean = 0
    sd = 1
    # Calculate number of random variates in experiment
    n = 2 ** i
    
    # Create distribution
    data_Gaussian = norm.rvs(size = n, loc = mean, scale = sd)
    # Create subplot
    ax = axs[row, col]
    ax.hist(data_Gaussian, bins = 50, color = "lightblue", density = True)
    title = 'Gaussian Distribution with ' + str(n) + " Random Variates"
    ax.set_title(title, size = 20)
    
    # Add probability density function
    rv_Gaussian = norm(loc = mean, scale = sd)
    x = np.linspace(-5, 5, 1000000)
    h = ax.plot(x, rv_Gaussian.pdf(x), lw = 2, color = "darkred")
    
    # Increment counter
    k += 1
    
# Add labels to axes
for ax in axs.flat:
    ax.set_xlabel('Value', fontsize = 20)
    ax.set_ylabel('Frequency', fontsize = 20)

# Hide X labels and tick labels for top plots and Y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Add title to entire figure
fig.suptitle("Gaussian Distribution with Varying Numbers of Random Variates", fontweight = "bold", size = 40)

plt.show()
fig.savefig('Plots/Gaussian_Distribution.png')
```

```python
## Part B: Simulate and plot linearly transformed Gaussian distribution

# Set up figure and axes for plotting
fig, axs = plt.subplots(3, 3, figsize = (30,30))
# Create counter to track number of iterations and calculate row and col values
k = 0
for i in range(10, 19):
    # Calculate row number
    row = k // 3
    # Calculate column number
    col = k % 3
    
    # Define parameters
    # Transformation: 5x + 10
    a = 5
    b = 10
    mean = 0
    sd = 1
    # Calculate number of random variates in experiment
    n = 2 ** i
    
    # Create distribution
    data_Gaussian = norm.rvs(size = n, loc = mean, scale = sd)
    # Apply linear transformation to distribution
    data_Gaussian = np.add(b, np.multiply(a, data_Gaussian))
    
    
    # Create subplot
    ax = axs[row, col]
    ax.hist(data_Gaussian, bins = 50, color = "lightblue", density = True)
    title = 'Gaussian Distribution with ' + str(n) + " Random Variates"
    ax.set_title(title, size = 20)
    
    # Add probability density function
    rv_Gaussian = norm(loc = b, scale = a)
    x = np.linspace(-10, 30, 1000000)
    rv_Gaussian_pdf = rv_Gaussian.pdf(x)
    h = ax.plot(x, rv_Gaussian_pdf, lw = 2, color = "darkred")
    
    # Increment counter
    k += 1
    
# Add labels to axes
for ax in axs.flat:
    ax.set_xlabel('Value', fontsize = 20)
    ax.set_ylabel('Frequency', fontsize = 20)

# Hide X labels and tick labels for top plots and Y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Add title to entire figure
fig.suptitle("Linearly Transformed (5x + 10) Gaussian Distribution with Varying Numbers of Random Variates", fontweight = "bold", size = 40)

plt.show()
fig.savefig('Plots/Gaussian_Distribution_LT.png')
```

```python
## Part C: Simulate and plot probability integral transformation of Gaussian distribution

# Set up figure and axes for plotting
fig, axs = plt.subplots(3, 3, figsize = (30,30))
# Create counter to track number of iterations and calculate row and col values
k = 0
for i in range(10, 19):
    # Calculate row number
    row = k // 3
    # Calculate column number
    col = k % 3
    
    # Define parameters
    mean = 0
    sd = 1
    start = mean
    end = sd
    # Calculate number of random variates in experiment
    n = 2 ** i
    
    # Draw from uniform distribution
    data_uniform = uniform.rvs(size = n, loc = start, scale = end)
    # Apply probability integral transformation
    data_Gaussian = norm.ppf(data_uniform, loc = mean, scale = sd)
    
    
    # Create subplot
    ax = axs[row, col]
    ax.hist(data_Gaussian, bins = 50, color = "lightblue", density = True)
    title = 'Gaussian Distribution with ' + str(n) + " Random Variates"
    ax.set_title(title, size = 20)
    
    # Add probability density function
    rv_Gaussian = norm(loc = mean, scale = sd)
    x = np.linspace(-5, 5, 1000000)
    rv_Gaussian_pdf = rv_Gaussian.pdf(x)
    h = ax.plot(x, rv_Gaussian_pdf, lw = 2, color = "darkred")
    
    # Increment counter
    k += 1
    
# Add labels to axes
for ax in axs.flat:
    ax.set_xlabel('Value', fontsize = 20)
    ax.set_ylabel('Frequency', fontsize = 20)

# Hide X labels and tick labels for top plots and Y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Add title to entire figure
fig.suptitle("Probability Integral Transformation of Gaussian Distribution with Varying Numbers of Random Variates", fontweight = "bold", size = 40)

plt.show()
fig.savefig('Plots/Gaussian_Distribution_PIT.png')
```

```python
## Part C: Simulate and plot probability integral transformation of exponential distribution

# Set up figure and axes for plotting
fig, axs = plt.subplots(3, 3, figsize = (30,30))
# Create counter to track number of iterations and calculate row and col values
k = 0
for i in range(10, 19):
    # Calculate row number
    row = k // 3
    # Calculate column number
    col = k % 3
    
    # Define parameters
    loc = 0
    scale = 1
    start = loc
    end = scale
    
    # Calculate number of random variates in experiment
    n = 2 ** i
    
    # Draw from uniform distribution
    data_uniform = uniform.rvs(size = n, loc = start, scale = end)
    # Apply probability integral transformation
    data_exp = expon.ppf(data_uniform, loc = loc, scale = scale)
    # Create subplot
    ax = axs[row, col]
    ax.hist(data_exp, bins = 50, color = "lightblue", density = True)
    title = 'Exponential Distribution with ' + str(n) + " Random Variates"
    ax.set_title(title, size = 20)
    
    # Add probability density function
    rv_exp = expon(loc = loc, scale = scale)
    x = np.linspace(0, 12, 100000)
    h = ax.plot(x, rv_exp.pdf(x), lw = 2, color = "darkred")
    
    # Increment counter
    k += 1
    
# Add labels to axes
for ax in axs.flat:
    ax.set_xlabel('Value', fontsize = 20)
    ax.set_ylabel('Frequency', fontsize = 20)

# Hide X labels and tick labels for top plots and Y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Add title to entire figure
fig.suptitle("Probability Integral Transformation of Exponential Distribution with Varying Numbers of Random Variates", fontweight = "bold", size = 40)

plt.show()
fig.savefig('Plots/Exponential_Distribution_PIT.png')
```
